# Data cleaning

```python
# Delete duplicate rows
dataset.drop_duplicates(inplace=True)
```

```python
# Transfer "year" to datetime format
dataset['year'] = pd.to_datetime(dataset['year'], format='%Y')
```

```python
# Delete rows with nan-value in "year" column
dataset = dataset.dropna(subset=['year'])
```

```python
# Delete rows with less than 0 and more than 1 million in "price" column
dataset = dataset[(dataset['price'] > 0) & (dataset['price'] < 1_000_000)]
```

```python
# Delete rows with less than 1-1-1900 and equal or more than 31-12-2024 in "year" column
dataset = dataset[(dataset['year'] > '1900-01-01') & (dataset['year'] <= '2024-12-31')]
```

```python
# Fill nan-value with "unknown" in "manufacturer" column
dataset['manufacturer'] = dataset['manufacturer'].fillna('Unknown')
```

```python
# Fill nan-value with "unknown" in "fuel" column
dataset['fuel'] = dataset['fuel'].fillna('Unknown')
```

```python
# Fill nan-value with "unknown" in "condition" column
dataset['condition'] = dataset['condition'].fillna('Unknown')
```

```python
# Fill nan-value with "unknown" in "transmission" column
dataset['transmission'] = dataset['transmission'].fillna('Unknown')
```

```python
# Fill nan-value with "unknown" in "type" column
dataset['type'] = dataset['type'].fillna('Unknown')
```

```python
# Fill nan-value with "unknown" in "paint_color" column
dataset['paint_color'] = dataset['paint_color'].fillna('Unknown')
```

```python
# Fill nan-value with "unknown" in "cylinders" column
dataset['cylinders'] = dataset['cylinders'].fillna('Unknown')
```

```python
# Fill nan-value with "unknown" in "model" column
dataset['model'] = dataset['model'].fillna('Unknown')
```

```python
# Fill nan-value with "unknown" in "title_status" column
dataset['title_status'] = dataset['title_status'].fillna('Unknown')
```

```python
# Fill nan-value with "unknown" in "drive" column
dataset['drive'] = dataset['drive'].fillna('Unknown')
```

```python
# Fill nan-value with median in "odometer" column
median_odometer = dataset['odometer'].median()
dataset['odometer'] = dataset['odometer'].fillna(median_odometer)
```

```python
# Drop the non-necessaire column
dataset.drop(columns=['id', 'url', 'region_url', 'image_url', 'description', 'county', 'VIN'], inplace=True)
```

```python
# Drop the non-necessaire column
dataset.drop(columns=['size'], inplace=True)
```

```python
# Drop the non-necessaire column
dataset.drop(columns=['lat', 'long'], inplace=True)
```

5. Which regions have the highest and lowest average car prices?

```python
price_by_region = dataset.groupby('region')['price'].mean().sort_values(ascending=False)
price_by_region
highest_avg_price_region = price_by_region.idxmax()
lowest_avg_price_region = price_by_region.idxmin()
print(f"Region with the highest average price: {highest_avg_price_region} ({price_by_region.max():,.2f} USD)")
print(f"Region with the lowest average price: {lowest_avg_price_region} ({price_by_region.min():,.2f} USD)")
```
